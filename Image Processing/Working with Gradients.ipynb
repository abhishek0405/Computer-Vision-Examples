```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showimg(img):
    plt.figure(figsize=(7,7))
    plt.imshow(img,cmap='gray')
```

```python
sudokuimg = cv2.imread('../DATA/sudoku.jpg',0)
```

```python
showimg(sudokuimg)
```

```python
ret,thres1 = cv2.threshold(sudokuimg,130,255,cv2.THRESH_BINARY_INV)
```

```python
showimg(thres1)
```

```python
sobelx = cv2.Sobel(sudokuimg,cv2.CV_64F,1,0,ksize=5)
showimg(sobelx)
```

```python
sobely = cv2.Sobel(sudokuimg,cv2.CV_64F,0,1,ksize=5)
showimg(sobely)
```

```python
blended_img = cv2.addWeighted(sobelx,0.95,sobely,0.5,0)
showimg(blended_img)
```

```python
ret,thresh2 = cv2.threshold(blended_img,250,255,cv2.THRESH_BINARY)
showimg(thresh2)
```

```python
gblurimg = cv2.GaussianBlur(blended_img,(5,5),35)
```

```python
showimg(gblurimg)
```

```python
ret,thresh3 = cv2.threshold(gblurimg,140,255,cv2.THRESH_BINARY)
np.power(thresh3,5)
showimg(thresh3)
```

```python
gblurimg2 = cv2.GaussianBlur(sudokuimg,(5,5),35)
showimg(gblurimg2)
```

```python
thresh4 = cv2.adaptiveThreshold(gblurimg2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,5,8)
showimg(thresh4)
```
