```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def showimg(img):
    plt.figure(figsize=(7,7))
    plt.imshow(img)
```

```python
rainbowbgr = cv2.imread('../DATA/rainbow.jpg')
rainbowrgb = cv2.cvtColor(rainbowbgr,cv2.COLOR_BGR2RGB)

```

```python
def plot_hist(img):
    color=('b','g','r')
    for i,color in enumerate(color):
        hist = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist)
    
```

```python
showimg(ranbowrgb)
```

```python
plot_hist(rainbowbgr)
```

```python
gorrilabgr = cv2.imread("../DATA/gorilla.jpg")
gorilargb = cv2.cvtColor(gorrilabgr,cv2.COLOR_BGR2RGB)
```

```python
showimg(gorilargb)
```

```python
plot_hist(gorrilabgr)
```
