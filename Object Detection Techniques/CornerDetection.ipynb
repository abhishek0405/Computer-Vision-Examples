```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Using harrisCorner Detection
chess = cv2.imread('../DATA/chessboard_mat.jpg');
showchess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
plt.imshow(showchess)
```

```python
chessbw = cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
new = cv2.cornerHarris(chessbw,2,3,0.04)
```

```python
plt.imshow(new)
```

```python
new1 = cv2.dilate(new,None)
```

```python
plt.imshow(new1)
```

```python
showchess[new1>0.02*new1.max()] = [255,0,0]
plt.imshow(showchess)
```

```python
chess = cv2.imread('../DATA/chessboard_mat.jpg')
```

```python
#Shi tomashi algorithm
corners = cv2.goodFeaturesToTrack(chessbw,15,0.01,10)
corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(chess,(x,y),6,(255,0,0),-1)
```

```python
plt.imshow(chess)
```
